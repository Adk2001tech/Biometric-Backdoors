```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
# adv_cen_loss
# target_cen_loss
loss0= np.load('/content/drive/MyDrive/Github/Biometric/Benchmark0.npy', allow_pickle=True).item()
loss_rdm= np.load('/content/drive/MyDrive/Github/Biometric/Benchmark_RDM.npy', allow_pickle=True).item()
```

```
plt.figure(figsize=(20, 8))
plt.title('Random Perturbation ->-> Adversarial Feature Space', size= 20)
plt.plot(loss0['adv_cen_loss'], linestyle='--', marker='o', label='Normal Loss')
plt.plot(np.array(loss_rdm['adv_cen_loss'])/10, linestyle='--', marker='+', label='RDM Loss/10')
plt.legend()
plt.grid()
plt.show()
```

```
plt.figure(figsize=(20, 8))
plt.title('Adversarial Centered Perturbation ->-> Traget Feature Space', size= 20)
plt.plot(loss0['target_cen_loss'], linestyle='--', marker='o', label='Normal Loss')
plt.plot(np.array(loss_rdm['target_cen_loss'])/1e4, linestyle='--', marker='+', label='RDM Loss/1e4')
plt.legend()
plt.grid()
plt.show()
```
